**Name: Telco Customer Churn Dataset**

Location: This dataset is available in Kaggle at https://www.kaggle.com/datasets/yeanzc/telco-customer-churn-ibm-dataset

Organization that generated: IBM generated this dataset for analytical purposes based on a fictional Telco company and a list of customers labeled with churn status.

**Content of the dataset**

It contains 7,043 records with a total of 33 columns.

The most relevant columns are:

CustomerID & the Churn Label (Yes/No), which allow us to differentiate two types of customers, those who remain in the compañy and those who left.

Also there are some other columns like:

Gender (male/female), Dependents (Yes/No) to know if the customer lives with other people where he is responsible, Tenure Months to indicate how many months the client was with the company services.

And there are a few columns to indicate the customer services contracted during his time like: Phone Services, Multiple Lines, Internet Service, Online Security, etc.

There are also some columns about the type of contract of the customer: Contract (monthly, annual, etc), Payment Method (deposit, credit card, mailed check, etc), Monthly Charge and Total Charges.

There are some additional columns for Geolocation which are very important like: State, City, Country, Zip Code, Longitude & Latitude.
